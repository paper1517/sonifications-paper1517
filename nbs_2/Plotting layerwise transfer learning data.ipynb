```python
import numpy as np
import seaborn as sns
import pandas as pd
import pickle
```

```python
with open("/media/user/nvme/contrastive_experiments/fsd50k_layerwise_data.pickle", "rb") as fd:
    fsd50k = pickle.load(fd)
```

```python
with open("/media/user/nvme/contrastive_experiments/speechcommands_layerwise_data.pickle", "rb") as fd:
    speechcommands = pickle.load(fd)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="whitegrid", palette="deep", font='Fira Mono', font_scale=0.75)
```

```python
fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(15,5))

colours = sns.color_palette("deep", 4)
ax1 = sns.barplot(x='layer', y='val_mAP', hue='model_type_alt', data=fsd50k['layerwise_df'], ci="sd", ax=axs[0])

ax1.axhline(fsd50k["contrastive_fc_val"], ls='--', color=colours[0])
ax1.axhline(fsd50k["supervised_fc_val"], ls='--', color=colours[1])
ax1.axhline(fsd50k["contrastive_full_val"], ls='--', color=colours[2])
ax1.axhline(fsd50k["supervised_full_val"], ls='--', color=colours[3])


ax1.legend(loc="upper left", fontsize=8)
ax1.set_title("Layer-wise FSD50K validation performance")


ax2 = sns.barplot(x='layer', y='val_acc', hue='model_type_alt', data=speechcommands['layerwise_df'], ci="sd", ax=axs[1])
ax2.axhline(speechcommands["contrastive_fc_val"], ls='--', color=colours[0])
ax2.axhline(speechcommands["supervised_fc_val"], ls='--', color=colours[1])
ax2.axhline(speechcommands["contrastive_full_val"], ls='--', color=colours[2])
ax2.axhline(speechcommands["supervised_full_val"], ls='--', color=colours[3])
ax2.legend(loc="upper left", fontsize=8)
ax2.set_title("Layer-wise Speech Commands performance")

plt.tight_layout()
plt.savefig("/home/user/Desktop/layerwise_performance_final.png", dpi=300, bbox_inches='tight')
```
